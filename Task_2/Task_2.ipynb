# Data Cleanising ( Data Quality Issues and cleaning )

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values
```

```python
# Drop rows with any missing values
df2 = df.dropna()

# Drop columns with 
threshold = 0.9 # 90% Threshold
df3 = df.dropna(thresh=df.shape[0] * threshold, axis =1)
df3.head()
```

```python
# Fill missing values in numeric columns with mean
numeric_col = ['CONFIRMED_STOREYS', 'CONFIRMED_UNITS']
df[numeric_col] = df[numeric_col].fillna(df[numeric_col].mean())
```

```python
# Fill missing values in categorical columns with mode
cat_col = ['PROPERTY_TYPE', 'WARD', 'WARDNAME']
df[cat_col] = df[cat_col].fillna(df[cat_col].mode().iloc[0])
```

```python
# Check for missing values
test = df.isnull().sum()
print(test)
```

```python
df_cleaned = df.dropna()
```

```python
# Check the final result
df_cleaned.isnull().sum()
```

```python
# Handling Outliers
from scipy.stats import zscore

num_col = ['SCORE']
z_score = df_cleaned[num_col].apply(zscore)
z_score.head()
```

```python
# Define a threshold for outliers detection 
threshold = 3
df_outliers = df_cleaned[(z_score < threshold).all(axis=1)]
df_outliers.head()
```

```python
# Standardizing Categorical and Numerical Variables

# convert categorical variables to lowercase
categorical_colmn = ['PROPERTY_TYPE', 'WARDNAME']
for column in categorical_colmn:
    df_cleaned[column] = df_cleaned[column].str.lower()
```

```python
# Normalize numerical variables using min-max scaling
from sklearn.preprocessing import MinMaxScaler

num_col = ['CONFIRMED_STOREYS', 'CONFIRMED_UNITS']
scaler = MinMaxScaler()
df_cleaned[num_col] = scaler.fit_transform(df_cleaned[num_col])
```

```python
# Creating binary feature indicating if the building is old or new based on year built
df_cleaned['IS_NEW_BUILDING'] = (df_cleaned['YEAR_BUILT'] > 2000).astype(int)
```

```python
# Extracting the month and year from the "EVALUATION_COMPLETED_ON" column
df_cleaned['EVALUATION_MONTH'] = pd.to_datetime(df_cleaned['EVALUATION_COMPLETED_ON']).dt.month
df_cleaned['EVALUATION_YEAR'] = pd.to_datetime(df_cleaned['EVALUATION_COMPLETED_ON']).dt.year
```

```python
# Agg by propert type
property_type = df_cleaned.groupby('PROPERTY_TYPE')['SCORE'].mean()
property_type
```

```python
# Agg by Ward
ward_stats = df_cleaned.groupby('WARD').agg({'SCORE': 'mean', 'PROPERTY_TYPE': 'count'})
ward_stats.head()
```

```python
# Agg by Year Built
year_stats = df_cleaned.groupby('YEAR_BUILT')['SCORE'].mean()
year_stats.head()
```
